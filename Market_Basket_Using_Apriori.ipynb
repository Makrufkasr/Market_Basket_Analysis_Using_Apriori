```python
#import data
product_name = pd.read_csv('https://docs.google.com/spreadsheets/d/1Tm7U4nczdAe45xbEPwxTRrNM_GM-hlsoBllcfZAZTGM/export?format=csv&gid=6207456')
product_name = pd.read_csv('')
product_name.head()
```

```python
print(len(product_name['user_id'].unique()))
print(len(product_name['product_name'].unique()))
```

```python
#import data
cluster = pd.read_csv('https://docs.google.com/spreadsheets/d/11YFZierNfCj2oQ64xcLQmID76qrKJMzdUcXd1mZ0cqI/export?format=csv&gid=1619293991')
cluster.head()
```

```python
df_merge = product_name.merge(cluster, how='left', left_on=['sessions_id','user_id'], right_on=['sessions_id','user_id'])
print(df_merge.shape)
df_merge.head()
```

```python
grouped_products = df_merge.groupby('sessions_id')['product_name'].apply(list).reset_index(name='product_list')
print(grouped_products['product_list'])
```

Cluster 0

```python
df_cluster0 = df_merge[df_merge['cluster'] == 0]
df_cluster0 = df_cluster0.loc[:,['user_id','product_name']]
print(df_cluster0.shape)
df_cluster0.head()
```

```python
transactions = [a[1]['product_name'].tolist() for a in list(df_cluster0.groupby(['user_id']))]#Combing All items in list Format for Each Member
transactions
```

#Importing and setting the apriori options


1.   Minimum support 0.002
2.   Minimum confidence 0.05
3.   Minimum Lift 3 
4.   Lenght the number of items




```python
from apyori import apriori ## Importing apriori package
rules = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.05, min_lift = 3, min_length = 2, max_length = 2) ## Model Creation
```

```python
results = list(rules)
```

Transform Result

```python
## Creating user-defined function for arranging the results obtained from model into readable format

def inspect(results):
    x         = [tuple(result[2][0][0])[0] for result in results]
    y         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(x, y, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['X', 'Y', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame.nlargest(n=20, columns="Lift") ## Showing best possible scenarios
```

```python
resultsinDataFrame.nlargest(n=20, columns="Support")
```

```python
resultsinDataFrame.nlargest(n=20, columns="Confidence")
```
